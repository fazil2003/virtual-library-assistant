```python
##### SORTED AND REVERSED LIST FUNCTION #####
def Sort(sub_li): 
    return(sorted(sub_li, key = lambda x: x[1], reverse=True))
```

```python
##### SORTED AND REVERSED THE DICTIONARY INTO LIST #####
query="artifigkdial inteljhdvigence books for new person"

appended_list=[]

for ind in df.index: 
    score=fuzz.token_set_ratio(query,df['name'][ind])
    value=[df['name'][ind],score, df['id'][ind]]
    appended_list.append(value)
appended_list=Sort(appended_list)

for i in appended_list:
    print(i[0]," - ",i[1]," - ",i[2])

#print(appended_list)
```
