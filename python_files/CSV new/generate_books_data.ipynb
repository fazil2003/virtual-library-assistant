## ADD COLUMN ID IN DATAFRAME

```python
import pandas as pd
df=pd.read_csv('books.csv')
```

```python
import numpy as np
book_id=np.arange(start=1, stop=2383, step=1)
```

```python
book_id.shape
```

```python
df['id']=pd.Series(book_id)
```

```python
### REORDER COLUMNS IN PANDAS DATAFRAME ###
df = df[["id","title","description","genre","image_link","book_rating","ratings_count","author","author_rating","year","edition","impressions","clicks"]]
```
